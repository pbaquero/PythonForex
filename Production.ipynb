```python
%%time
'''Creating empty base and final arrays'''

arr_data=df_tick_data[:100] #need to 1)find right datatype (smallest) 2)optimize structure, and 3)determine memory usage
#print(test_data)

arr_base_len=len(arr_data.index)
arr_final_len= len(arr_data.index)-1
#arr_final_len= (len(arr_data.index)*(len(arr_data.index)-1))//2


arr_base=np.zeros(arr_base_len, dtype={'names':('TTime','Bid','Ask'),
                          'formats':('datetime64[ms]','f8','f8')})

arr_final=np.zeros(arr_final_len, dtype={'names':('PTime','STime', 'PTBid','PTAsk', 'STBid', 'STAsk', 'Profit', 'SBid_Ask_Spread','PBid_Ask_Spread','TimeDelta'),
                          'formats':('datetime64[ms]','datetime64[ms]', 'f8','f8','f8','f8','f8','f8','f8','timedelta64[ms]')})

#test_arr_1=np.zeros((arr_len,4))
#test_arr_1[0]


#print(test_arr_1[1][1])
#print(test_arr_1[1].shape)
#print(test_arr_1)
#test_arr_1

'''
print(test_arr.shape)

print(test_data.count)
print(test_data.shape)
print(len(test_data.index))
'''
print(arr_final.dtype.itemsize)
print(arr_final.nbytes)
print(arr_final)
```
