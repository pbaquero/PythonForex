```python
pd.options.mode.chained_assignment = None

df_plot_col = USD_JPY_df.drop(['Complete','Time', 'Volume'], axis=1).copy()
df_plot_sub = df_plot_col[:100]
df_plot_sub['BidAvg']= df_plot_sub[['Bid_Open','Bid_High','Bid_Low','Bid_Close']].mean(axis=1)
df_plot_sub['AskAvg']= df_plot_sub[['Ask_Open','Ask_High','Ask_Low','Ask_Close']].mean(axis=1)

sma = [5,10,20,30,40,50]
lma = [10,20,50,100,250,500]

for x in sma:
    df_plot_sub['Bid_SSMA'+str(x)]= df_plot_sub['Bid_Close'].rolling(x).mean()
    df_plot_sub['Bid_SEMA'+str(x)]= pd.Series.ewm(df_plot_sub['Bid_Close'],span = x).mean()
    
for x in lma:
    df_plot_sub['Bid_LSMA'+str(x)]= df_plot_sub['Bid_Close'].rolling(x).mean()
    df_plot_sub['Bid_LEMA'+str(x)]= pd.Series.ewm(df_plot_sub['Bid_Close'],span = x).mean()

'''
df_plot_sub['Bid_SMA5']= df_plot_sub['Bid_Close'].rolling(5).mean()
df_plot_sub['Bid_EMA5']= pd.Series.ewm(df_plot_sub['Bid_Close'],span = 5).mean()
df_plot_sub['Bid_SMA10']= df_plot_sub['Bid_Close'].rolling(10).mean()
df_plot_sub['Bid_EMA10']= pd.Series.ewm(df_plot_sub['Bid_Close'],span = 10).mean()
'''

display(df_plot_sub)
print(df_plot_sub.columns)
```

```python
%matplotlib qt

#fig = plt.figure()
#ax1 = plt.subplot2grid((1,1), (0,0))

fig = plt.figure()
ax1 = fig.add_subplot(111) # in this function, first number 1 represent the height,
# second is for width and third is the plot number
ax2 = fig.add_subplot(111)

df_plot_sub['ohlc_date'] = df_plot_sub.index.map(mdates.date2num)

pre_ohlc = df_plot_sub[['ohlc_date','Bid_Open','Bid_High','Bid_Low','Bid_Close']]

ohlc = list(pre_ohlc.itertuples(index = False, name = None))

#need to set width to number of datapoints within a day (24*60*12)
candlestick_ohlc(ax1, ohlc, width=0.6/(24*60*12), colorup='#77d879', colordown='#db3f3f')

for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)

ax1.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d %H:%M:%S'))
ax1.xaxis.set_major_locator(mticker.MaxNLocator(10))
ax1.grid(True)

df_plot_sub.MA10.plot(ax=ax2)
df_plot_sub.EMA10.plot(ax=ax2)

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Test')
plt.legend()
plt.subplots_adjust(left=0.09, bottom=0.20, right=0.94, top=0.90, wspace=0.2, hspace=0)
plt.show()
```
